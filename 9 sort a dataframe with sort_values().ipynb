## sort the data frame 

```python
#  sort by one column
df_exams.sort_values(by='math score')
```

```python
# sort by descxending by ine column 
df_exams.sort_values('math score',ascending=False)
```

```python
df_exams
```

```python
#sort descending by multiple columns 
# here we can only obtain the copy of data frame its not  orginal data frame
#using inplace=true we can update the orgginal data frame 
df_exams.sort_values(['math score','reading score'],ascending=False,inplace=True)
```

```python
#now our original data frame is updated 
df_exams 
```

```python
# sort descending with a key function 
df_exams.sort_values('race/ethnicity',ascending=True,key=lambda col:col.str.lower())
```
