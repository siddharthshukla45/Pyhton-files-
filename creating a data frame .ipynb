# creating a data frame from an array 


```python
import pandas as pd 
import numpy as np 
```

### 1.1Option 1 

```python
#creating an array 
data=np.array([[1,4,],[2,5],[3,6]])
```

```python
#creating a dataframe 
df=pd.DataFrame(data,index=['row1 ','row2','row3'],
            columns=['col1','col2'])
```

```python
#showing the dataframe 
df
```

#  2.creating a data frame using a dictionary 

```python
#lists used for this example
states=['Utter Pradesh','Bihar','Himachal pradesh','jammu kashmir']
population=[230000000,12304505,10203456,30456790]
```

```python
#storing list within a dictionary 
dict_states={'States':states,'Population':population}
```

```python
#creating a data frame 
df.population=pd.DataFrame(dict_states)
```

```python
#showing the data frame 
df.population
```
